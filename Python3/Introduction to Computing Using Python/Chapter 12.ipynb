```python
import sqlite3

def webData(databaseName, url, linkList, wordFreq):
    con = sqlite3.connect(databaseName)
    cur = con.cursor()
    cur.execute("CREATE TABLE Keywords (Url text, Word text, Freq text)")
    cur.execute("CREATE TABLE Hyperlinks (Url text, Link text)")
    
    for word in wordFreq.keys():
        insert = (url, word, wordFreq[word])
        cur.execute("INSERT INTO Keywords VALUE (?, ?, ?)", insert)
    
    for link in linkList:
        insert = (url, link)
        cur.execute("INSERT INTO Hyperlinks VALUE (?, ?)", insert)
        
    con.commit()
    con.close()
    
def freqSearch(db):
    word = input('Enter keyword: ')
    
    con = sqlite3.connect(db)
    cur = con.cursor()
    
    stringFormat = '{:30} {:5}'
    print(stringFormat.format('URL', 'FREQ'))
    
    query = "SELECT * FROM Keywords WHERE Word = ?"
    var = (word,)
    
    cur.execute(query, var)
    
    result = cur.fetchall()
    
    for row in result:
        print(stringFormat.format(row[0], row[2]))
    
    con.close()
```

```python
from string import punctuation

class SeqMapReduce(object):
    def __init__(self, mapper, reducer):
        self.mapper = mapper
        self.reducer = reducer
    def process(self, data):
        intermediate1 = [self.mapper(x) for x in data]
        intermediate2 = partition(intermediate1)
        return [self.reducer(x) for x in intermediate2]

def occurence(word):
    return [(word, 1)]

def partition(intermediate1):
    dct = {}
    
    for lst in intermediate1:
        for key, value in lst:
            if key in dct:
                dct[key].append(value)
            else:
                dct[key] = [value]
    
    return dct.items()

def occurenceCount(keyVal):
    return (keyVal[0], sum(keyVal[1]))

def getWordsFromFile(file):
    infile = open(file)
    content = infile.read()
    infile.close()
    
    transTable = str.maketrans(punctuation, ' '*len(punctuation))
    content = content.translate(transTable)
    
    res = set()
    for word in content.split():
        res.add((word, file))
    return res

def getWordIndex(keyVal):
    return keyVal

def getChars(word):
    'Mapper'
    return [(char, word) for char in word]

def getCharIndex(charPair):
    'Reducer'
    return charPair

yo = ['ant', 'bee', 'cat', 'dog', 'eel']
mp = SeqMapReduce(getChars, getCharIndex)
mp.process(yo)
```

```python
from time import time
from math import ceil, sqrt

def prime1(n):
    for i in range(2,n):
        if n%i == 0:
            return False
    return True

def prime2(n):
    if n%10 in [0,2,4,5,6,8]:
        return False
    elif sum([int(i) for i in str(n)])%3 == 0:
        return False
    
    for j in range(7,n//6):
        if n%j == 0:
            return False
    
    return True

def nFirstPrimes(minNum, m):
    primeList = []
    for i in range(minNum, m+1):
        if prime1(i):
            primeList.append(i)
    return primeList
            
def prime3(n, pList, largerPrime):
    if n%10 in [0,2,4,5,6,8]:
        return False
    elif sum([int(i) for i in str(n)])%3 == 0:
        return False
    
    for prime in pList:
        if n%prime == 0:
            return False
  
    for j in range(largerPrime, n//largerPrime, 2):
        if n%j == 0:
            return False
    
    return True

def prime4(n, pList, largerPrime):
    
    for prime in pList:
        if n%prime == 0:
            return False
  
    for j in range(largerPrime, n//largerPrime, 2):
        if n%j == 0:
            return False
    
    return True

def prime5(n):
    'returns True if integer n is prime, False otherwise'
    if n == 2:
        return True
    if n < 2 or n % 2 == 0:
        return False
    upper = ceil(sqrt(n))
    i = 3
    while i <= upper:
        if n%i == 0:
            return False
        i += 2
    return True

def startTest(start, rng):
    avg1 = []
    avg2 = []
    primeNumpy = []
    primeNum2 = []
    pList = nFirstPrimes(7,400)
    pList2 = nFirstPrimes(2,400)
    largerPrime = max(pList)
    for i in range(5):
        for n in range(start, start+rng):
            t1 = time()
            if prime5(n):
                primeNumpy.append(1)
            t2 = time()
            if prime4(n, pList2, largerPrime):
                primeNum2.append(1)
            t3 = time()
            avg1.append(t2-t1)
            avg2.append(t3-t2)

    print('First Function')
    print(sum(avg1))
    print(sum(primeNumpy))
    print('Second Function')
    print(sum(avg2))
    print(sum(primeNum2))

startTest(1000000, 100000)
```

```python
from multiprocessing import Pool

class MapReduce(object):
    'a parallel implementation of map reduce'
    
    def __init__(self, mapper, reducer, numProcs = None):
        'initialize map and reduce functions, and process pool'
        
        self.mapper = mapper
        self.reducer = reducer
        self.pool = Pool(numProcs)
        
    def process(self, data):
        'runs MapReduce on sequence data'
        
        intermediate1 = self.pool.map(self.mapper, data)
        intermediate2 = partition(intermediate1)
        return self.pool.map(self.reducer, intermediate2)

def primeFac(num):
    'return prime number of input number'
    
    res = set()
    
    for i in range(2, num+1):
        while num%i == 0:
            num = num/i
            res.add(i)
    
    return res

def partition(intermediate1):
    
    dct = {}    
    for lst in intermediate1:
        for key, value in lst:
            if key in dct:
                dct[key].append(value)
            else:
                dct[key] = [value]
    return dct.items()

# def primeMapper(lst):
    
#     res = []
    
#     for num in lst:
#         for prime in primeFac(num):
#             res.append((prime, num))
#     return res

def primeMapper(num):
    return [(prime, num) for prime in primeFac(num)]

def primeReducer(lst):
    return lst

yo = MapReduce(primeMapper, primeReducer)
yo.process([24, 15, 35, 60])
```
