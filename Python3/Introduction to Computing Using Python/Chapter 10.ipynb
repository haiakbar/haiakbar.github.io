```python
from turtle import Screen, Turtle
def koch(n):
    if n==0:
        return 'F'
    tmp = koch(n-1)
    return tmp+'L'+tmp+'R'+tmp+'L'+tmp

def drawKoch(n):
    s = Screen()
    t = Turtle()
    directions = koch(n)
    for move in directions:
        if move == 'F':
            t.forward(300/3**n)
        if move == 'L':
            t.lt(60)
        if move == 'R':
            t.rt(120)
    s.bye()

def snowflake(n):
    s = Screen()
    t = Turtle()
    directions = koch(n)
    for i in range(3):
        for move in directions:
            if move == 'F':
                t.forward(300/3**n)
            if move == 'L':
                t.lt(60)
            if move == 'R':
                t.rt(120)
        t.rt(120)
    s.bye()


snowflake(3)
```

```python
import time

def power(a,n):
    res = 1
    for i in range(n):
        res *= a
    return res

def rpower(a,n):
    if n==0:
        return 1
    tmp = rpower(a,n//2)
    
    if n%2==0:
        return tmp*tmp
    else:
        return a*tmp*tmp

def pow(a,n):
    return a**n

def power2(k):
    return power(2,k)

def rpower2(k):
    return rpower(2,k)

def pow2(k):
    return pow(2,k)

def timing(func, n):
    
    start = time.time()
    func(n)
    end=time.time()
    return end-start

def timingAnalysis(func, start, stop,inc,runs):
    for n in range(start,stop,inc):
        acc=0.0
        
        for i in range(runs):
            acc += timing(func,n)
            
        formatStr = 'Run time of {}({}) is {:.7f} seconds'
        print(formatStr.format(func.__name__,n,acc/runs))

timingAnalysis(power2,20000, 80000+1, 20000, 10)
timingAnalysis(rpower2,20000, 80000+1, 20000, 10)
timingAnalysis(pow2, 20000, 80000+1, 20000, 10)
```

```python
import time
import random

def search(lst, target, i,j):
    if i==j:
        return -1
    
    mid = (i+j)//2
    if lst[mid]==target:
        return mid
    if target < lst[mid]:
        return search(lst,target,i,mid)
    else:
        return search(lst,target,mid,j)
    
def binary(lst):
    target = random.choice(lst)
    return search(lst,target,0,len(lst))

def linear(lst):
    target = random.choice(lst)
    return lst.index(target)

def buildInput(n):
    lst = random.sample(range(2*n),n)
    lst.sort()
    return lst

def timing(func, n):
    funcInput = buildInput(n)
    start = time.time()
    func(funcInput)
    end=time.time()
    return end-start

def timingAnalysis(func, start, stop,inc,runs):
    for n in range(start,stop,inc):
        acc=0.0
        
        for i in range(runs):
            acc += timing(func,n)
            
        formatStr = 'Run time of {}({}) is {:.7f} seconds'
        print(formatStr.format(func.__name__,n,acc/runs))
        
timingAnalysis(linear, 200000, 1000000, 200000, 20)
timingAnalysis(binary, 200000, 1000000, 200000, 20)
```

```python
import random
import time

def buildList(n):
    res = []
    for i in range(n):
        res.append(random.choice(range(n**2)))
    return res

def timing(func, n):
    funcInput = buildList(n)
    start = time.time()
    func(funcInput)
    end=time.time()
    return end-start

def timingAnalysis(func, start, stop, inc, runs):
    for n in range(start,stop,inc):
        acc=0.0

        for i in range(runs):
            acc += timing(func,n)
            
        formatStr = 'Run time of {}({}) is {:.7f} seconds'
        print(formatStr.format(func.__name__,n,acc/runs))

def frequent(lst):
    lst.sort()
    currentLen=1
    longestLen=1
    mostFreq=lst[0]
    
    for i in range(1,len(lst)):
        if lst[i]==lst[i-1]:
            currentLen+=1
        else:
            if currentLen>longestLen:
                longestLen = currentLen
                mostFreq=lst[i-1]
            currentLen=1
    
    return mostFreq

def frequent2(lst):
    newLst = {}
    
    #construct dict based on list
    for item in lst:
        if item in newLst:
            newLst[item] += 1
        else:
            newLst[item] = 1
    
    
    
    keyLst = []
    
    #construct key list
    for key in newLst.keys():
        keyLst.append(key)
    
    mostFreq = keyLst[0]
    
    for i in range(1,len(newLst)):
        if newLst[keyLst[i]]<newLst[keyLst[i-1]]:
            if newLst[keyLst[i-1]]>newLst[mostFreq]:
                mostFreq = keyLst[i-1]
        else:
            if newLst[keyLst[i]]>newLst[mostFreq]:
                mostFreq = keyLst[i]
    
    #return item with the most occurences
    return mostFreq

timingAnalysis(frequent,2000,8000+1,2000,10)
timingAnalysis(frequent2,2000,8000+1,2000,10)
```

```python
def rearrange(word):
    chars = []
    for char in word:
        chars.append(char)
    return sorted(''.join(chars))

def anagram(filename, target):
    infile = open(filename,'r')
    text = infile.read()
    infile.close()
    
    text = text.split()
    
    target = rearrange(target)
    
    for word in text:
        if rearrange(word) == target:
            print(word)

anagram('words.txt', 'trace')
```

```python
def pascalLine(l):
    if l == 0:
        return [0]
    elif l == 1:
        return [1,1]
    else:
        prev = pascalLine(l-1)
        return [1]+[prev[i]+prev[i+1] for i in range(len(prev)-1)]+[1]
pascalLine(4)
```

```python
import os
signatures = {
    'Creeper':'ye8009g2h1azzx33',
    'Code Red':'99dh1cz963bsscs3',
    'Blaster':'fdp1102k1ks6hgbc'
}
def scan(pathname, signatures):
    if os.path.isfile(pathname):
        infile = open(pathname)
        content = infile.read()
        infile.close()
        
        for virus in signatures:
            if content.find(signatures[virus])>=0:
                print('{}, found virus {}'.format(pathname, virus))
        return
    for item in os.listdir(pathname):
        fullpath = os.path.join(pathname, item)
        
        scan(fullpath, signatures)
scan('test', signatures)
```

```python
from turtle import Screen, Turtle

def levy(n):
    if n == 1:
        return 'LFRFL'
    else:
        prev = levy(n-1)
        return 'L'+prev+'R'+prev+'L' 
def drawLevy(n):
    s = Screen()
    t = Turtle()
    directions = levy(n)
    for move in directions:
        if move == 'F':
            t.forward(200/n)
        if move == 'L':
            t.lt(45)
        if move == 'R':
            t.rt(90)
    s.bye()
    
drawLevy(8)
```

```python
def coins(n):
    if n == 8:
        return True
    elif n < 8:
        return False
    
    chck = n%10 == 0 or n%2 == 0 or n%3 == 0 or n%4 == 0
    if chck:
        if n%10 == 0 and coins(n-9) == True:
            return True 
        if n%2 == 0 and coins(n/2+1) == True:
            return True
        if n%3 == 0 and coins(n-7) == True:
            return True
        if n%4 == 0 and coins(n-6) == True:
            return True
    
    return False
    
coins(20)
```

```python
def largerValue(lst):
    lrg = max(lst)
    lst.remove(lrg)
    return lrg

def recReverse(lst):
    cont = []
    
    for i in range(len(lst)):
        lrg = largerValue(lst)
        cont.append(lrg)
    return cont

recReverse([1,2,3,4,5,6,7,8,9])
```

```python
from turtle import Turtle, Screen
s = Screen()
t = Turtle()
t.pensize(2)

def square(turtle, x, y, length):
    turtle.penup()
    turtle.goto((length/2)+x, (length/2)+y)
    turtle.setheading(180)
    turtle.pendown()
    for i in range(4):
        turtle.forward(length)
        turtle.lt(90)
        
def squares(turtle, x, y, length, n):
    if n == 1:
        square(turtle, x, y, length)
    else:
        square(turtle, x, y, length)
        
        corner = [((length/2)+x, (length/2)+y), 
                  (-(length/2)+x, (length/2)+y),
                  ((length/2)+x, -(length/2)+y),
                  (-(length/2)+x, -(length/2)+y)]
        
        for x, y in corner:
            squares(turtle, x, y, length/2.2, n-1)
            
squares(t, 0, 0, 200, 5)
s.bye()
```
