```python
from turtle import Turtle, Screen

class Disk(Turtle):
    'a tower of hanoi disk class'
    
    def __init__(self, n):
        'initialize disk n'
        Turtle.__init__(self, shape='square', visible=False)
        self.penup()
        self.sety(300)
        self.shapesize(1, 1.5*n, 2)
        self.fillcolor(1, 1, 1)
        self.showturtle()
    
class Peg(Turtle, list):
    'a tower of hanoi peg class, inherits from Turtle and list'
    pos = -200
    
    def __init__(self, n):
        'initializes a peg for n disks'
        
        Turtle.__init__(self, shape='square', visible=False)
        self.penup()
        self.shapesize(n*1.25, 0.75, 1)
        self.sety(12.5*n)
        self.x = Peg.pos
        self.setx(self.x)
        self.showturtle()
        Peg.pos += 200
    
    def push(self, disk):
        disk.setx(self.x)
        disk.sety(10+len(self)*25)
        self.append(disk)
    
    def popp(self):
        disk = self.pop()
        disk.sety(300)
        return disk
        
def moveDisk(fromPeg, toPeg):
    'moves top disk from fromPeg to toPeg'
    
    disk = fromPeg.popp()
    toPeg.push(disk)

def hanoi(n, peg1, peg2, peg3):
    'move n disks from peg1 to peg3 using peg2'
    
    #base case: n == 0. Do nothing
    
    if n > 0:
        hanoi(n-1, peg1, peg3, peg2)
        moveDisk(peg1, peg3)
        hanoi(n-1, peg2, peg1, peg3)

def play(n):
    screen = Screen()
    Peg.pos = - 200
    p1 = Peg(n)
    p2 = Peg(n)
    p3 = Peg(n)
    
    for i in range(n):
        p1.push(Disk(n-i))
    
    hanoi(n, p1, p2, p3)
    
    screen.bye()

play(4)
```

```python
from turtle import Turtle, Screen

class Line(Turtle):
    'a line bar'
    pos = -200
    
    def __init__(self, n):
        'initializes a peg for n disks'
        
        Turtle.__init__(self, shape='square', visible=False)
        self.penup()
        self.shapesize((n+1)*1.25, 0.5, 1)
        self.sety(12.5*n)
        self.x = Line.pos
        self.setx(self.x)
        self.showturtle()
        Line.pos += 30

def pattern(n):
    if n == 0:
        return str(0)
    
    return pattern(n-1) + str(n) + pattern(n-1)
        
def patternGUI(n):
    screen = Screen()
    Line.pos = -200
    
    for height in pattern(n):
        height = int(height)
        Line(height)
        
    
    
    screen.bye()
    
patternGUI(4)
```
